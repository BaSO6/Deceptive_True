## 4.1 Transformation stage

````python
import os
import pandas as pd
import librosa
import random
import numpy as np
import scipy.stats
import librosa.feature
import librosa
import numpy as np
import scipy.stats

# feature extraction function
def extract_audio_features(audio_path, duration=10):
    y, sr = librosa.load(audio_path, sr=None)

    # feature extraction
    frame_length = 2048
    hop_length = 512
    n_fft = min(1024, len(y))

    # time domain frequece
    short_time_energy = np.mean([sum(abs(y[i:i+frame_length]**2)) for i in range(0, len(y), hop_length)])
    zero_crossing_rate = np.mean(librosa.feature.zero_crossing_rate(y)[0])
    short_time_amplitude = np.mean([np.mean(abs(y[i:i+frame_length])) for i in range(0, len(y), hop_length)])
    rms = np.mean(librosa.feature.rms(y=y, frame_length=n_fft, hop_length=hop_length)[0])
    waveform_envelope = np.mean([max(abs(y[i:i+frame_length])) for i in range(0, len(y), hop_length)])
    time_domain_mean = np.mean(y)
    time_domain_variance = np.var(y)
    time_domain_kurtosis = scipy.stats.kurtosis(y)
    time_domain_skewness = scipy.stats.skew(y)
    time_domain_max = np.max(y)  # max
    time_domain_min = np.min(y)  # min
    time_domain_range = time_domain_max - time_domain_min  # range（max - min）
    time_domain_median = np.median(y)  # middle
    time_domain_std = np.std(y)  # variance
    time_domain_25_percentile = np.percentile(y, 25)  # 25%
    time_domain_75_percentile = np.percentile(y, 75)  # 75%

    # frequece domain
    spectral_centroid = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))
    spectral_bandwidth = np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))
    spectral_rolloff = np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))
    spectral_flatness = np.mean(librosa.feature.spectral_flatness(y=y, n_fft=n_fft, hop_length=hop_length))
    spectral_contrast = np.mean(librosa.feature.spectral_contrast(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))
    spectral_peak = np.max(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))  # 频谱峰值
    spectral_mean = np.mean(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))  # 频谱均值

    # other stastic feature
    spectral_std = np.std(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))  # 频谱标准差
    spectral_median = np.median(librosa.feature.spectral_centroid(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))  # 频谱中位数

    # MFCC feature
    mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13, n_fft=n_fft, hop_length=hop_length)
    mfcc_mean = np.mean(mfcc, axis=1)
    mfcc_std = np.std(mfcc, axis=1)  # MFCC的variance
    mfcc_max = np.max(mfcc, axis=1)  # MFCC max
    mfcc_min = np.min(mfcc, axis=1)  # MFCC min
    mfcc_range = mfcc_max - mfcc_min  # MFCC range=（max - min）

    # more frequecy domain feature
    chroma_stft = np.mean(librosa.feature.chroma_stft(y=y, sr=sr, n_fft=n_fft, hop_length=hop_length))  # 色度特征
    tonnetz = np.mean(librosa.feature.tonnetz(y=y, sr=sr))  # pitch（Tonnetz）

    # return all features
    return {
        "short_time_energy": short_time_energy,
        "zero_crossing_rate": zero_crossing_rate,
        "short_time_amplitude": short_time_amplitude,
        "rms": rms,
        "waveform_envelope": waveform_envelope,
        "time_domain_mean": time_domain_mean,
        "time_domain_variance": time_domain_variance,
        "time_domain_kurtosis": time_domain_kurtosis,
        "time_domain_skewness": time_domain_skewness,
        "time_domain_max": time_domain_max,
        "time_domain_min": time_domain_min,
        "time_domain_range": time_domain_range,
        "time_domain_median": time_domain_median,
        "time_domain_std": time_domain_std,
        "time_domain_25_percentile": time_domain_25_percentile,
        "time_domain_75_percentile": time_domain_75_percentile,
        "spectral_centroid": spectral_centroid,
        "spectral_bandwidth": spectral_bandwidth,
        "spectral_rolloff": spectral_rolloff,
        "spectral_flatness": spectral_flatness,
        "spectral_contrast": spectral_contrast,
        "spectral_peak": spectral_peak,
        "spectral_mean": spectral_mean,
        "spectral_std": spectral_std,
        "spectral_median": spectral_median,
        "mfcc_mean": mfcc_mean.tolist(),  
        "mfcc_std": mfcc_std.tolist(),
        "mfcc_max": mfcc_max.tolist(),
        "mfcc_min": mfcc_min.tolist(),
        "mfcc_range": mfcc_range.tolist(),
        "chroma_stft": chroma_stft,
        "tonnetz": tonnetz,
    }


# Save audio features to CSV
def process_wav_folder(wav_folder, output_folder):
    for subfolder_name in os.listdir(wav_folder):
        subfolder_path = os.path.join(wav_folder, subfolder_name)
        if os.path.isdir(subfolder_path):
            audio_features_list = []  
            
            for wav_file in os.listdir(subfolder_path):
                if wav_file.endswith(".wav"):
                    wav_path = os.path.join(subfolder_path, wav_file)
                    features = extract_audio_features(wav_path)

                    # Save the filenames with the feature
                    features['filename'] = wav_file
                    audio_features_list.append(features)
            
           
            if audio_features_list:
                features_df = pd.DataFrame(audio_features_list)
                output_file = os.path.join(output_folder, f"{subfolder_name}_features.csv")
                features_df.to_csv(output_file, index=False)
                print(f"Processed and saved {output_file}")

# Input and Output paths
wav_folder = "./MLEndDD_stories_small"  # Input
output_folder = "./Output"  # Output

# Create output folder
os.makedirs(output_folder, exist_ok=True)

# process wavs
process_wav_folder(wav_folder, output_folder)
````

- **Why Chosen**:
  - **Learns from Model Interactions**: Unlike voting, stacking allows the meta-classifier to learn complex patterns from the combined predictions of the base models.
  - **Improves Accuracy**: By weighting predictions appropriately, stacking can outperform individual models and simple voting classifiers.

Random Forest optimizes features using a combination of **random sampling** and **feature selection** techniques to build diverse decision trees. Here’s how it works:

---

### 1. **Feature Subset Selection at Each Split**
   - At each split in a decision tree, Random Forest selects a random subset of features (controlled by the `max_features` parameter).
   - This random selection ensures that each tree in the forest explores a different set of relationships between features and the target variable.
   - **Impact**:
     - Reduces overfitting by decorrelating the trees.
     - Focuses on the most informative features in different parts of the feature space.

---

### 2. **Bootstrap Sampling (Bagging)**
   - Each tree is trained on a bootstrap sample (a random sample with replacement) of the dataset.
   - This introduces variability in the training data for each tree, allowing the model to capture different aspects of the feature space.
   - **Impact**:
     - Ensures that some features (and their patterns) are prioritized in certain trees while ignored in others.

---

### 3. **Feature Importance Calculation**
   Random Forest calculates feature importance after training by evaluating how each feature contributes to reducing impurity (e.g., Gini impurity or entropy). This is typically done using two main approaches:
   - **Mean Decrease in Impurity (MDI)**:
     - Tracks the reduction in impurity (e.g., Gini index) whenever a feature is used for a split across all trees.
     - Features that consistently lead to large reductions in impurity are considered more important.
   - **Mean Decrease in Accuracy (MDA)**:
     - Measures the decrease in model accuracy when a feature's values are randomly permuted.
     - If permuting a feature significantly reduces accuracy, it indicates the feature is important.

---

### 4. **Bias-Variance Tradeoff**
   - Random Forest uses many weak learners (shallow decision trees) to reduce variance and overfitting.
   - By averaging predictions across multiple trees, it achieves a balance between fitting complex patterns and generalizing well.

---

### 5. **Tuning Parameters for Feature Optimization**
   - **`max_features`**: Controls the number of features considered at each split. Smaller values focus on a more diverse set of trees, while larger values allow the trees to use more of the available data.
   - **`max_depth`**: Limits the depth of each tree to prevent overfitting.
   - **`min_samples_split`** and **`min_samples_leaf`**: Control the minimum number of samples required to make splits or be a leaf node, reducing noisy splits.

---

### Advantages of Random Forest for Feature Optimization:
1. **Robust to Irrelevant Features**:
   - Even if the dataset contains many irrelevant features, Random Forest tends to ignore them since they don’t contribute to impurity reduction or accuracy.
2. **Handles High Dimensionality**:
   - Works well with datasets containing many features, as the random selection at each split reduces the computational cost.
3. **Feature Ranking**:
   - Outputs feature importance scores, allowing for interpretability and further refinement of the model.

---

### Summary
Random Forest optimizes features by:
1. Randomly sampling features for splits (decorrelating trees).
2. Evaluating feature contributions via impurity reduction or accuracy impact.
3. Averaging predictions from multiple trees to mitigate overfitting.

### Constant-Q Transform (CQT) vs. Mel Spectrograms

Both **Constant-Q Transform (CQT)** and **Mel Spectrograms** are time-frequency representations of audio, but they are optimized for different tasks and have distinct characteristics. Here's a detailed comparison:

---

### 1. **Frequency Bins and Resolution**
| **Aspect**          | **Constant-Q Transform (CQT)**                        | **Mel Spectrogram**                          |
|----------------------|-------------------------------------------------------|----------------------------------------------|
| **Frequency Bins**   | Logarithmically spaced (e.g., equal-width bins in octaves). | Linearly spaced in the Mel scale.            |
| **Frequency Range**  | More bins for lower frequencies, fewer bins for higher frequencies. | Uniform resolution across the Mel scale.     |
| **Frequency Resolution** | High resolution at low frequencies, low resolution at high frequencies. | Resolution is uniform, independent of frequency. |

**Key Difference**:  
- CQT aligns better with human auditory perception, which has a logarithmic sensitivity to pitch. This makes it better suited for harmonic and tonal analysis.
- Mel spectrogram provides a smoother representation and is computationally efficient, making it ideal for general-purpose audio analysis.

---

### 2. **Time Resolution**
| **Aspect**         | **Constant-Q Transform (CQT)**                        | **Mel Spectrogram**                          |
|---------------------|-------------------------------------------------------|----------------------------------------------|
| **Time Resolution** | High resolution for higher frequencies, lower resolution for lower frequencies. | Uniform resolution across all frequencies.   |
| **Temporal Characteristics** | Captures rapid changes in higher frequencies but may lose detail in lower frequencies. | Balances time and frequency resolution uniformly. |

**Key Difference**:  
- CQT sacrifices temporal resolution for lower frequencies in favor of frequency accuracy.
- Mel spectrogram has uniform time resolution, making it better for tasks that require both frequency and temporal detail, like speech processing.

---

### 3. **Applications**
| **Aspect**                  | **Constant-Q Transform (CQT)**               | **Mel Spectrogram**                          |
|-----------------------------|-----------------------------------------------|----------------------------------------------|
| **Best Use Cases**           | Music analysis, harmonic analysis, pitch tracking. | Speech recognition, emotion detection, general audio tasks. |
| **Suitability for Speech**   | Good for tonal nuances (e.g., emotional cues). | Better for phonetic and linguistic features. |

**Key Difference**:  
- CQT excels in tasks involving tonal or harmonic structures, such as detecting pitch or emotional cues.
- Mel spectrogram is better for phoneme recognition and tasks requiring uniform time-frequency resolution.

---

### 4. **Computation**
| **Aspect**                  | **Constant-Q Transform (CQT)**               | **Mel Spectrogram**                          |
|-----------------------------|-----------------------------------------------|----------------------------------------------|
| **Complexity**               | More computationally intensive due to logarithmic bin spacing. | Faster to compute with simpler algorithms.   |
| **Scalability**              | Slower for large datasets.                   | Scales well for real-time and large datasets. |

**Key Difference**:  
- CQT is computationally heavier but provides richer harmonic details.
- Mel spectrogram is efficient and widely used in real-time systems.

---

### 5. **Visualization**
| **Aspect**                  | **Constant-Q Transform (CQT)**               | **Mel Spectrogram**                          |
|-----------------------------|-----------------------------------------------|----------------------------------------------|
| **Appearance**               | Displays logarithmically spaced frequencies, emphasizing harmonic content. | Smoother and evenly distributed along the frequency axis. |
| **Interpretability**         | Better for analyzing harmonic relationships and tonal changes. | Easier to interpret for general speech and sound analysis. |

**Key Difference**:  
- CQT often highlights harmonics and overtones, making it ideal for analyzing music or emotional speech.
- Mel spectrograms are easier to interpret visually for tasks like phoneme recognition.
